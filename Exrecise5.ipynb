```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from math import sin, exp

def Plot(f):
    return (sin(f-2)**2)*exp(-f**2)
f = np.linspace(0,2,100)
plotF = np.vectorize(Plot)

plt.figure(figsize=(10,10))
plt.plot(f, plotF(f),label='Plot(f)')
plt.grid()
plt.tick_params(labelsize=30)
plt.legend(loc='best',fontsize=10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as sts

figure, first_x = plt.subplots();
index_z = np.random.randn(10000)
index_x = np.linspace(-5, 5, 1000)
z = sts.gaussian_kde(index_z)
e_z = z.pdf(index_x)
first_x.hist(index_z, bins=25, density=True)
first_x.plot(index_x, e_z, lw=1.7)
plt.show()
```
